```python
import pandas as pd
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import xgboost as xgb
sns.set_style('darkgrid')
pd.set_option('display.max_columns', None)
import datetime, warnings, scipy
warnings.filterwarnings("ignore")

import pickle
```

```python
# Read in datasets of combined flight and weather data
# All2017Flight_Weather.csv is the file name from SettingUpCSVTrainingYearly.ipynb
df17 = pd.read_csv('All2017Round2_InDelay.csv', index_col=None)
df17.head(3).append(df17.tail(3))
df17 = df17.loc[:, ~df17.columns.str.contains('^Unnamed')]

df16 = pd.read_csv('All2016Round2_InDelay.csv', index_col=None)
df16.head(3).append(df16.tail(3))
df16 = df16.loc[:, ~df16.columns.str.contains('^Unnamed')]

df15 = pd.read_csv('All2015Round2_InDelay.csv', index_col=None)
df15.head(3).append(df15.tail(3))
df15 = df15.loc[:, ~df15.columns.str.contains('^Unnamed')]

df14 = pd.read_csv('All2014Round2_InDelay.csv', index_col=None)
df14.head(3).append(df14.tail(3))
df14 = df14.loc[:, ~df14.columns.str.contains('^Unnamed')]
```

```python
# Combine years 2014-2017 to form training dataset
df1417 = pd.concat([df17, df16, df15, df14])
```
